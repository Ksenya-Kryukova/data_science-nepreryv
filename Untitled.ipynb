Задание 1
Получить интуицию, на что влияют параметры распределений. Построить по 5 функций плотностей с разными параметрами для каждого распределения: нормальное, экспоненциальное, Стьюдента на одном графике.

```python
mu1 = 1.0
mu2 = 1.5
mu3 = 2.0
mu4 = 2.5
mu5 = 3.0
sigma1 = 0.5
sigma2 = 0.75
sigma3 = 0.25
sigma4 = 0.15
sigma5 = 0.9
```

```python
# зададим нормально распределенную случайную величину
norm_rv1 = sts.norm(loc=mu1, scale=sigma1)
norm_rv2 = sts.norm(loc=mu2, scale=sigma2)
norm_rv3 = sts.norm(loc=mu3, scale=sigma3)
norm_rv4 = sts.norm(loc=mu4, scale=sigma4)
norm_rv5 = sts.norm(loc=mu5, scale=sigma5)
```

```python
x = np.linspace(0, 5, 100)

pdf1 = norm_rv1.pdf(x)
pdf2 = norm_rv2.pdf(x)
pdf3 = norm_rv3.pdf(x)
pdf4 = norm_rv4.pdf(x)
pdf5 = norm_rv5.pdf(x)

plt.plot(x, pdf1)
plt.plot(x, pdf2, color = 'red')
plt.plot(x, pdf3, color = 'black')
plt.plot(x, pdf4, color = 'yellow')
plt.plot(x, pdf5, color = 'green')

plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.show()
```

Мат ожидание мю определяет центр плотности распределения, а сигма задает насколько она узкая или широкая, т.е. чем больше сигма, тем функция более широкая и низкая визуально

```python
lam1 = 0.1
lam2 = 0.2
lam3 = 0.3
lam4 = 0.5
lam5 = 0.9
```

```python
# зададим экспоненциально распределенную случайную величину
expon_rv1 = sts.expon(lam1)
expon_rv2 = sts.expon(lam2)
expon_rv3 = sts.expon(lam3)
expon_rv4 = sts.expon(lam4)
expon_rv5 = sts.expon(lam5)
```

```python
x = np.linspace(0, 5, 1000)

pdf1 = expon_rv1.pdf(x)
pdf2 = expon_rv2.pdf(x)
pdf3 = expon_rv3.pdf(x)
pdf4 = expon_rv4.pdf(x)
pdf5 = expon_rv5.pdf(x)

plt.plot(x, pdf1)
plt.plot(x, pdf2, color = 'red')
plt.plot(x, pdf3, color = 'black')
plt.plot(x, pdf4, color = 'yellow')
plt.plot(x, pdf5, color = 'green')

plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.show()
```

чем больше лямбда, тем дальше пик и функция более крутая (резкая)

```python
n1 = 5
n2 = 10
n3 = 20
n4 = 50
n5 = 1000
```

```python
# зададим распределенную случайную величину по Стьюденту
t_rv1 = sts.t(n1)
t_rv2 = sts.t(n2)
t_rv3 = sts.t(n3)
t_rv4 = sts.t(n4)
t_rv5 = sts.t(n5)
```

```python
x = np.linspace(-5, 5, 1000)

pdf1 = t_rv1.pdf(x)
pdf2 = t_rv2.pdf(x)
pdf3 = t_rv3.pdf(x)
pdf4 = t_rv4.pdf(x)
pdf5 = t_rv5.pdf(x)

plt.plot(x, pdf1)
plt.plot(x, pdf2, color = 'red')
plt.plot(x, pdf3, color = 'black')
plt.plot(x, pdf4, color = 'yellow')
plt.plot(x, pdf5, color = 'green')

plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.show()
```

чем больше n, тем функция уже, пик выше, и сама она более крутая

Известно, что светофор горит 54 секунды красным, 4 секунды жёлтым и 28 секунд зелёным. Посчитайте, какое количество информации несёт сообщение о цвете светофора в текущий момент. А что будет в случае 28, 28 и 28 секунд для каждого цвета?

```python
import math
```

Всего 54 + 4 + 28 = 86 с
Вероятность красного = 54/86
Вероятность оранжевого = 4/86
Вероятность зеленого = 28/86

```python
Ired = math.log2(86/54)
print(Ired)

Ior = math.log2(86/4)
print(Ior)

Igr = math.log2(86/28)
print(Igr)
```

Логично, ведь оранжевый меньше всех по времени, поэтому количество информации данное сообщение несет больше всех


В случае, когда все по 28 с, т.е. получается , что все события равновероятно, формула упрощается до I=log2(N), где N - количество событий
